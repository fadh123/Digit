```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head(3)
tips_by_day = tips[['tip','day']].groupby(['day'], as_index=False).mean()
tips_by_day
plt.figure(figsize=(8,6))
plt.pie(tips_by_day['tip'], labels=tips_by_day['day'])
plt.axis('equal')
plt.title('Pie Chart: Average Tips Everyday', size=15)
plt.show()
plt.figure(figsize=(8,6))
plt.pie(tips_by_day['tip'],
        labels = tips_by_day['day'],
        autopct = '%.2f',
        explode = (0, 0.2, 0, 0.1)
        )
plt.axis('equal')
plt.title('Pie Chart: Average Tips Everyday', size=17)
plt.show()
```
